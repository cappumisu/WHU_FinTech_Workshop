## 1.涉及文献概要
1.Beaver（1968）通过成交量和收益残差平方来衡量市场反应，表明盈利公告对市场确实有显著影响，体现在成交量的增加和价格波动性的提高上。  
2.Ball and Brown（1968）研究了是财务报告中的盈利公告对股票价格的影响。发现股票市场对于即将公布的盈利信息反应灵敏，许多信息在正式公布前已经被市场所消化，盈利公告提供的信息占比很小。  
3.Bamber et al.（2000）认为后续研究者对Beaver（1968）中证据的解释过度泛化。文章对Beaver（1968）进行了复现和扩展研究，发现对样本选择等实验设计进行了修改后结果并不稳健，大多数个股的盈利公告并未引起异常的价格反应。研究表明如果采取不同的实验设计，盈利公告的信息内容并不那么明显，甚至可能不存在。  
4.Ball and Shivakumar（2008）量化了盈利公告在向股票市场提供新信息方面的相对重要性，并认为盈利公告相对而言并不重要。他们指出盈利公告的低频率和会计收入主要基于过去信息的事实，与市场预期的高频率更新相比，盈利公告提供的新信息有限。

## 2.主要内容介绍
1.本次实验同样使用NYSE上市的企业，但不同于Beaver (1968) ，本次使用了**全部**企业数据。   
2.主要关注财务年度结束于**2010/01/01-2019/12/31**之间的盈利报告。  
3.主要关注**年度**财务报告的公布。  
4.通过盈利公告日前后市场调整回报标准差（sd_ret_mkt）、平均成交量比（mean_rel_vol）和市场调整回报的平均绝对偏差（mad_ret_mkt）来映射市场对盈利公告的反应。  
5.采用事件发生前后n周作为non-event weeks，Beaver (1968)采用前后**8周**，本次实验采用**20个交易日**。  

## 3.Discussion1

### 3.1 Beaver（1968）和Ball and Brown（1968）的研究问题有何不同？如果存在重叠，您认为哪篇论文提供的证据更有利？
+ Beaver（1968）通过分析盈利公告前后的交易量和价格波动性来研究市场对盈利公告的反应。文章提出的方法可以直接观察到投资者对盈利信息的反应，从交易行为（交易量增加）和市场波动性两个维度捕捉市场情绪的变动。而Ball and Brown(1968)讨论了盈利公告内具体的会计数字（net income, EPS）怎样反映市场和个别企业的财务表现，市场是如何预期这些数字并据此调整股票价格的，并且探讨了盈利公告的及时性。二者不同点在于前者强调市场对盈利公告发布的即时反应，而后者强调盈利公告中的会计信息在更宽时间范围内对市场贡献的信息价值。
+ 两篇文献都研究了盈利公告发布对股票收益率的影响。哪篇提供的证据更有利取决于需要关注的问题。如果关注短期内市场对盈利公告的即使反应，则Beaver（1968）更有利；如果关注长期内盈利公告中会计信息的效率和可用性，则Ball and Brown（1968）更有利。

### 3.2 Beaver（1968）中使用的数据（如样本期）与Ball and Brown（1968）中使用的数据有何不同？为什么会存在这些差异？
+ 不同点：  
（1）Beaver（1968）使用的是1961-1965年的周度数据，Ball and Brown（1968）使用的是1946-1966年的年度数据。  
（2）Beaver（1968）和Ball and Brown（1968）的样本选择标准不同。  
+ 原因：研究目标不同。  
（1）Beaver（1968）侧重于市场短期反应，选择较短时间长度的周度数据；Ball and Brown（1968）侧重于盈利的长期市场预期和信息内容，选择了较长时间的年度数据。  
（2）Beaver（1968）的研究需要严格地剥离除盈利公告外的其他事件对成交量和股价的影响，样本选择标准大多数针对这一点进行筛选；Ball and Brown（1968）的样本选择标准使财政年结束日期统一，有助于简化分析，并确保结果不会因不同财政年结束日期的公司可能存在的季节性或日历效应而产生偏差。


### 3.3 Beaver（1968）样本选择标准的理由看起来合理吗？你认为这些是在得到结果之前制定的吗？为什么？这些决定（样本选择标准）是在研究过程中的哪个阶段做出的是否重要？  
+  样本选择标准：  
（1）公司必须在Compustat数据库上；  
（2）公司必须是纽约证券交易所（NYSE）的成员；  
（3）财政年度必须在12月31日之外的其他日期结束；  
（4）在盈利公告的同一周内没有宣布股息；  
（5）在围绕盈利公告的17周期间内没有股票拆分；  
（6）每年在《华尔街日报》上的新闻公告少于20次。
+ 不一定。可能是在得到初步结果后补充制定。
+ 重要，但并非必须在实验开始前确定所有标准。  
（1）理想情况下，样本选择标准的设定通常应该在研究设计阶段完成，因为样本选择直接影响到研究有效性和结论可推广性。在研究开始之前确定这些标准有助于确保研究过程的透明性和公正性，防止研究者基于数据分析后的结果来选择对其最有利的分析方法，导致结果不客观。  
（2）然而，在研究过程中，可能会发现初步的实验设计存在一些缺陷，这时调整样本选择标准以改善研究设计是合理的。这种调整有助于提高研究的准确性和相关性。但这也可能带来风险，比如放大了那些符合我们预期的个别样本的影响（Bamber et al. ，2020）。所以在进行样本选择时需要平衡预先设定的标准和根据初步结果进行的调整，以确保研究结果的广泛适用性和客观性。  
（3）Beaver（1968）中的样本选择标准主要是为了减少其他信息影响，专注于研究盈利公告对市场反应的影响。他选择了非12月31日财年末的公司来减少市场广泛性信息的影响，同时进一步剔除了包含其他信息的样本，以尝试隔离盈利公告作为独立信息来源的效果。这些标准有助于使研究聚焦于盈利公告对成交量量和价格波动性的影响，从而使研究目标更加明确和集中。因此Beaver（1968）制定的样本选择标准虽然可能是在实验过程中发现实验设计的缺陷后才完善的，但理由对于文章特定的研究目标和研究背景而言仍是合理的。尽管Beaver（1968）的样本选择标准存在局限性，文章结论的可推广性也因此受到质疑，但这并不意味着其样本选择不合理。相反，它突显了在设计研究时对样本选择标准进行仔细考虑的重要性，以及在解释研究结果时持谨慎态度的必要性。


### 3.4 对于Beaver（1968）的研究问题，股价和成交量哪个变量更好？这两个变量都使用有帮助吗？  
+ 成交量更好。股价反映了市场整体预期的变化，而成交量反映了个别投资者预期的变化。新信息的出现可能对市场的影响是中立的，即不改变市场整体预期，但很可能改变个人的预期。在这种情况下，股价不会出现反应，但成交量会出现反应。
+ 有帮助。在盈利公告公布周观测到股价或成交量的明显变动都能说明盈利公告给市场提供了新的信息，而且能帮助更深入地理解市场对盈利公告的反应。同时使用股价和成交量可以提供一个更全面的分析框架。股价变动显示市场对信息的整体评价，而成交量变化揭示了市场对该信息的反应强度和参与程度。二者都使用有助于识别单一变量难以发现的模式，例如一个盈利公告引起了股价的轻微变动但伴随着大幅的成交量增加可能表明市场对该公告看法存在分歧。


### 3.5 Beaver（1968）比较了event week和non-event week的成交量（volume）与其波动。Ball and Shivakumar（2008）测量了股票市场中盈利公告的公布在提供新的市场信息方面的相对重要性，认为盈利公告相对而言并不重要。如果成交量是信息量的测量方法，那么Beaver（1968）的figure 1在盈利公告那几周的信息量比例说明了什么？这与 Beaver（1968）的结果和Ball and Shivakumar（2008）的讨论相吻合吗？  


+ figure 1说明：  
（1）在公告前的8周内，成交量低于正常水平（non-event week均值），表明投资者可能推迟了股票交易直到盈利公告发布。  
（2）成交量在公告发布周出现明显峰值，说明盈利公告的发布促进了交易，对市场具有影响力，成交量随后的迅速回落说明投资者对新信息反应迅速。  
（3）在盈利公告发布的4周后，成交量略高于正常水平，表明投资者可能花时间对盈利数据进行了更深入的评估。  
+ 相吻合。两篇文献都观察到了盈利报告窗口期内成交量的异常增加。Beaver (1968）强调了盈利公告在短期内对市场的影响，而Ball and Shivakumar (2008)侧重于盈利公告在给市场提供的新信息相对占比较小，并没有否认盈利公告含有新信息。两篇论文的结论在各自的背景下都可能是正确的：Beaver (1968）强调了盈利公告期间短期内信息激增的情况，而Ball和Shivakumar（2008）考虑的是其在持续的信息流中盈利公告的相对重要性。

### 3.6 Beaver（1968）在p.77（residual volume analysis）和 pp.81-82（return residual analysis）讨论了统计显著性。为什么Beaver（1968）要讨论这些？如果现在写这篇论文，你现在会如何衡量统计显著性？

+ 可以用t检验来检验图t=0处的预测误差均值是否显著大于0，用t检验的p值来衡量统计显著性。


### 3.7 Beaver（1968）中的主要分析以图片形式提供。虽然自1968年以来绘图成本大幅下降，但我们现在通常不会看到以绘图形式呈现的初步分析，为什么会出现这种情况？
+ Beaver（1968）中对图表的大量使用是因为受到当时技术和方法论的限制。但现在越来越多的统计分析方法可以对数据进行更复杂、准确的分析，仅仅依靠图片得出结论相较而言不够严谨。因此图片更多作为辅助工具，帮助进行直觉性的初步探索和可视化理解，它们补充了统计分析，而不是作为主要的分析方法。

### 4.2 单个盈利公告
CTI BioPharma Corporation，一家生物制药公司，GVKEY为064515，该公司宣布了截至2017年12月31日的年度盈利，公告发布日期为2018年3月4日。

```python
single_event = earn_annc_dates.query("gvkey == '064515' and rdq == @pd.to_datetime('2018-03-04')")
single_event
```

```python
ccm_link = ccmxpf_lnkhist[(ccmxpf_lnkhist['linktype'].isin(["LC", "LU", "LS"])) &
                          (ccmxpf_lnkhist['linkprim'].isin(["C", "P"]))]
ccm_link = ccm_link.rename(columns={"lpermno": "permno"})
ccm_link['linkenddt'] = ccm_link['linkenddt'].fillna(ccm_link['linkenddt'].max())
ccm_link = ccm_link.reset_index(drop=True)
ccm_link['permno'] = ccm_link['permno'].str.replace('.0', '').str.pad(width=5, fillchar='0')
ccm_link
```

```python
single_event_link = pd.merge(single_event, ccm_link, how='inner', on='gvkey')
single_event_link = single_event_link.query("rdq >= linkdt and rdq <= linkenddt")[['gvkey', 'datadate', 'permno']]
single_event_link
```

```python
# 交易日编号
trading_dates = dsi[['date']].sort_values('date').reset_index(drop=True)
trading_dates['td'] = trading_dates.index + 1
trading_dates
```

```python
# 日期对应的交易日编号
min_date = trading_dates['date'].min()
max_date = trading_dates['date'].max()

annc_dates = (pd.DataFrame({'date': pd.date_range(min_date, max_date)})
              .merge(trading_dates, how='left', on='date')
              .bfill() # 非交易日的公告对应下一交易日的市场反应
              .rename(columns={'date': 'annc_date'}))
annc_dates['td'] = annc_dates['td'].astype(int)
annc_dates
```

```python
days_before = 20 # Beaver(1968) 采用 n=8（周）
days_after = 20

# 以交易日编号td为媒介计算窗口起止日期，起止日期只保留交易日
single_event_window = (pd.merge(single_event, 
                                annc_dates, 
                                how='left', 
                                left_on='rdq',
                                right_on='annc_date')
                       .drop(columns=['annc_date']))
single_event_window['start_td'] = single_event_window['td'] - days_before
single_event_window['end_td'] = single_event_window['td'] + days_after
single_event_window.rename(columns={'td': 'event_td'}, inplace=True)
single_event_window = (pd.merge(single_event_window, 
                                trading_dates , 
                                how='inner',
                                left_on='start_td',
                                right_on='td')
                       .rename(columns={'date': 'start_date'})
                       .drop(columns=['td']))
single_event_window = (pd.merge(single_event_window, 
                                trading_dates , 
                                how='inner',
                                left_on='end_td',
                                right_on='td')
                       .rename(columns={'date': 'end_date'})
                       .drop(columns=['td']))
single_event_window
```

### 5.1 After reading Bamber et al.（2000）, do the reasons given by Beaver（1968）for his sample selection criteria still seem reasonable? Why or why not?  
+ 仍是合理的。原因如下：  
（1）Beaver（2008）的目标是研究盈利公告对市场反应的影响。通过选择非12月31日财年末的公司，并限制样本中的新闻数量，以减少其他信息对市场反应的干扰，使研究更加集中在盈利公告的影响上。这种方法有助于提高分析的清晰度和专注度。  
（2）Bamber et al.（2000）通过复制Beaver（2008）的研究，并使用不同的样本（Fortune 200），发现只有一小部分盈利公告引起了市场的显著反应，表明样本选择标准可能会影响研究结论的普遍性和可推广性。  
（3）Bamber et al.（2000）的研究揭示了Beaver（2008）样本选择标准上的局限性，但这并不意味着文章的样本选择不合理。在不过度推广和泛化文章结论的前提下，Beaver（2008）的样本选择标准是符合文章特定的研究目标和研究背景的。

+ Beaver（1968）的样本选择标准和理由：  
    > **1) Firm must be on the Compustat tape**  
    理由：覆盖了超过90%市场市值的可以公开持有的股票，数据易获得  
    > **2) firm must be a member of the new york stock exchange**  
    理由：数据易获得  
    >Compustat:  
      -是一个全面的财务数据库, 收录公司的营运及财务状况资料库. 涵盖130多个国家, 超过 112,000上市/已退市公司  
      -the COMPUSTAT tapes contain incorrect data and that studies based on the tapes may be biased because they include only large, well known firms or only firms that have been in existence over a specified period.  
      -the selection of only large firms may tend to bias the research sample in favor of firms that are more successful and possibly less risky than the general population  
    >NYSE:  
      -400 shareholders   
      -1.1 million publicly held shares  
      -A minimum share price of $4.00.
-$40 million market value of those shares for IPO companies or $100 million for companies looking to list or transfer existing securities.
-Total pre-tax income earnings of at least $10 million in for the last three fiscal years (although there are some alternatives to this requirement). 
     缺点：larger firms--greater flow    
    > **3) fiscal year must end on a date other than Dec.31st**  
    理由：防止一段时间内的信息太多，导致的整体市场波动（如果没有本限制，会导致2、3、4月数据过多，因为有2/3的Compustat公司的会计年末都是12/31.）  
    缺点：在NYSE和Compustat外，会筛选出很多的小规模企业，其占据了一定的市场份额；零售和食物业在样本中比例变高（后续研究显示，零售业和食物业的价格、交易量波动都更小）  
    > **4) no dividends were announced in the same week as the earnings announcement**    
    > **5) no stock splits were announced during the 17 week period surrounding the announcement of earnings**  
    理由：减少其他事件对earning announcement带来的不确定性影响  
    > **6) less than 20 news announcements per year appearing in the Wall Street Journal**  
    理由：便于分析  
    缺点：使得研究结果倾向于表明收益公告具有较高的信息内容，因为选择的样本倾向于那些较少受到媒体关注的公司，这些公司的收益公告更可能被市场视为新信息  

只要这种标准是事前确定的，那么研究结果适用的范围就很容易被确定。  


+ Bamber et al. (2000)认为：Beaver's (1968)的（3）、（6）样本选择导致了【smaller-than-average firms】。    
Beaver的样本只占NYSE 5%的市值，其研究范围（根据以下文献）可能会引起更明显的市场反应  
Smith & Pourciau (1988),Grant (1980) and Shores (1990) and Atiase (1985) and Bamber (1987)：price and volume reactions to small firms' earnings announcements exceed the reactions to larger firms' announcements；reaction to earnings announcements is a decreasing function of the number of WSJ announcements.

### 5.2 Bamber等人（2000）复制并扩展了Beaver（1968）的研究。为什么复制研究部分很重要？Bamber等人（2000）还可以考虑哪些研究扩展方法？将扩展限制在Beaver（1968）的样本范围、可用数据和方法上是否合理？你如何看待“第一块砖（即Beaver，1968）影响了整个会计研究史”的说法？
+ 包含Beaver（1968）复现重要的原因：  
（1）验证结果的稳健性。通过复现Beaver（1968）的研究，Bamber et al.（2000）能够验证原始研究结果的可靠性和稳健性，从而确保研究结果不是偶然得出的，而是可以在相似条件下重复得到的。  
（2）验证研究方法的有效性。复现Beaver（1968）能帮助评估Beaver（1968）使用的研究方法和样本选择标准的有效性，有助于识别原始研究中可能存在的局限性或偏差。在不同时间段复现Beaver（1968）的研究，能够评估原结论在变化的市场环境中是否仍然成立。  
（3）为扩展研究提供基础。通过首先复现再进行拓展研究，Bamber et al.（2000）可以将新结论和原结论作比较，了解研究方法改变带来的影响。
+ 已使用的替代方法：  
（1）使用Fortune 200的公司；  
（2）样本期用1962-1966代替原本的1961-1965；  
（3）取消了一年内20条WSJ报道的限制；  
（4）按公告周U-statistics在公告窗口期U-statistics的排名划分为1-17组，用每组所占比例作为研究指标，反映盈利公告能显著提供信息的公司占比。  

可以考虑的其他替代方法：  
（1）扩大样本。可以扩大样本，包括不同交易所、行业或国家的公司。  
（2）检验其他变量。可以探索其他财务披露或公司事件的信息内容，而不仅仅关注盈利公告。  
（3）使用不同的技术方法。可以采用不同的统计或计量经济方法论来分析数据，例如使用机器学习技术或考虑日内数据的事件研究法。
+ 是否要限制在Beaver的研究框架内：  
优点：允许与原始研究进行直接比较，确保任何发现的差异是由于市场或会计实践的变化，而不是方法论的差异。  
缺点：它可能限制了研究的普遍性，并忽略了金融市场和会计研究中的重要发展。例如随着时间的推移，市场环境、会计准则和技术发展都可能影响盈利公告对市场的影响。因此，考虑这些变化对扩展研究也十分重要。    

+ 早期研究确实会对学术界产生重大影响，尽管Beaver在文章中提到了自己研究的缺陷和结论适用范围，也有很多文章指出此结论在Beaver样本外的企业不成立，但Beaver的结论（Earning Announcements 会给市场传递新信息）还是成了公知信息。  
However, it's important to recognize that while seminal studies lay the groundwork, the "wall" of research is continuously evolving, with new bricks being added and old ones reexamined in light of new evidence and methodologies.
